```python
empty_list = []
for i  in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit(df)
    score = km.score(df)
    empty_list.append(score)
```

```python
print(empty_list)
```

```python
empty_list = np.round(np.abs(empty_list))
```

```python
empty_list
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (15,10))
number_of_clusters = list(range(1,11))
plt.plot(number_of_clusters, empty_list, marker = "*", color = "Black", mec = "green", dash_capstyle = "projecting"  );
plt.xlabel("Number of Clusters");
plt.ylabel("SSD");
plt.grid()
```
